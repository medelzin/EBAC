````python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import numpy as np
````

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
df = sns.load_dataset("tips")
df
````

````python
df['net_bill'] = df['total_bill']-df['tip']
df['tip_pct'] =  df['tip'] / df['net_bill'] 
df.head()
````

````python
reg = smf.ols('tip_pct ~ net_bill', data = df).fit()
reg.summary()
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
reg1 = smf.ols('tip_pct ~ np.log(net_bill)', data = df).fit()
reg1.summary()
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
reg2 = smf.ols('tip_pct ~ np.power(net_bill,2)', data = df).fit()
reg2.summary()
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
reg3 = smf.ols('np.log(tip_pct) ~ net_bill', data = df).fit()
reg3.summary()
````

````python
df['pred_tip_pct'] = np.exp(reg3.fittedvalues)
df[['pred_tip_pct','tip_pct']].corr().iloc[0,1]**2
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
reg4 = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', data = df).fit()
reg4.summary()
````

````python
df['pred_tip_pct'] = np.exp(reg4.fittedvalues)
df[['pred_tip_pct','tip_pct']].corr().iloc[0,1]**2
````
