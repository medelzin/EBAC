````python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import numpy as np
````

### 1. Explorando a base de gorjetas

a. Na aula fizemos uma regressão de ```tip``` explicada por ```net_bill```. Carregue os pacotes e a base, crie as variáveis necessárias como na aula, e reproduza esta regressão.

b. Reproduza também o gráfico de dispersão dessas duas variáveis.

````python
df = sns.load_dataset("tips")
df
````

````python
df['net_bill'] = df['total_bill']-df['tip']
df['tip_pct'] =  df['tip'] / df['net_bill'] 
df.head()
````

````python
reg = smf.ols('tip ~ net_bill', data = df).fit()
reg.summary()
````

````python

sns.regplot(data=df,x='net_bill',y = "tip")
````

### 2. Mudança de perspectiva

Agora ajuste um modelo de regressão de ```tip_pct``` por ```net_bill```. Construa o gráfico de dispersão correspondente.

````python
reg2 = smf.ols('tip_pct ~ net_bill', data = df).fit()
reg2.summary()
````

````python
dados_novos = df
dados_novos
````

````python
dados_novos["pred"] = reg.predict(dados_novos)
dados_novos["pred"]
````

````python
sns.scatterplot(data=dados_novos,x='net_bill',y = "pred")
plt.plot([0,50], [0.1,0.1], linewidth=2,label="linha 10%")
````
