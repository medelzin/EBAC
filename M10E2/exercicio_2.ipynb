````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv("previsao_de_renda.csv")
df.head()
````

````python
df.drop("Unnamed: 0",axis=1, inplace=True)
df.head()
````

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
df_qualitativas = df[["posse_de_veiculo","posse_de_imovel","renda"]]
````

````python
grafico_veiculos = sns.pointplot(x="posse_de_veiculo",y="renda",data=df_qualitativas,dodge=True, ci=90)
````

````python
grafico_imoveis = sns.pointplot(x="posse_de_imovel",y="renda",data=df_qualitativas,dodge=True, ci=90)
````

````python
ax1 = sns.pointplot(x="posse_de_imovel",y="renda",hue="posse_de_veiculo",data=df_qualitativas,dodge=True, ci=90)
````

````python
ax2 = sns.pointplot(x="posse_de_veiculo",y="renda",hue="posse_de_imovel",data=df_qualitativas,dodge=True, ci=90)
````

Ambos os dados são significativos quando o assunto é afetar renda, porém é claro que possuir veículo é mais significativo.
